```python
import matplotlib.pyplot as plt
import numpy as np
from math import factorial
```

```python
N_space = np.linspace(1,10,11)
N = 10
T = 5
Delta = 1
```

```python
def temporal_pert_space(n, delta):
    return n**delta/factorial(delta)*(1+2*delta/n)**n
```

```python
temporal_pert_space(784, 2)
```

```python
temporal_pert_space(784, 3)
```

```python
temporal_pert_space(784, 4)
```

```python
temporal_pert_space(784, 4) / temporal_pert_space(784, 3) *1.7e4
```

```python
def ratio_func(t, alpha, n):
    return ((t**(alpha*t))/(1+2*alpha*t))**n

def ratio_by_n(n):
    return ratio_func(5, 0.1, n)
```

```python
plt.plot(ratio_by_n(N_space))
```

```python
ratio_func(T, 0.06, N_space)
```
